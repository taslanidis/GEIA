```python
import os
import re

# Dictionaries to store the results
MLC = {}
MSP = {}

# Set the directory path containing .out files
folder_path = 'output/'

# Regex patterns to match the required metrics
pattern_pre = r"micro precision_score on token level:\s([0-9]*\.[0-9]+)"
pattern_rec = r"micro recall_score on token level:\s([0-9]*\.[0-9]+)"
pattern_f1 = r"micro f1_score on token level:\s([0-9]*\.[0-9]+)"

# Iterate over each file in the specified folder
for filename in os.listdir(folder_path):
    # Check if the file is a .out file
    if filename.endswith('.out'):
        file_path = os.path.join(folder_path, filename)
        name = filename.split('_')
        if len(name) == 6:
            _, _, _, dat, victim, modelout = name
            model, _ = modelout.split("-")
        elif len(name) == 7:
            _, _, _, dat, victim1, victim2, modelout = name
            victim = victim1 + victim2
            model, _ = modelout.split("-")
            
        # Initialize nested dictionary for MLC or MSP if not already
        if model == "NN":
            if dat not in MLC:
                MLC[dat] = {}
            if victim not in MLC[dat]:
                MLC[dat][victim] = {}
        elif model == "RNN":
            if dat not in MSP:
                MSP[dat] = {}
            if victim not in MSP[dat]:
                MSP[dat][victim] = {}
        
        # Open and read the file
        with open(file_path, 'r') as file:
            for line in file:
                # Search for each metric in the line
                match_pre = re.search(pattern_pre, line)
                match_rec = re.search(pattern_rec, line)
                match_f1 = re.search(pattern_f1, line)
                
                # Save the matched precision score
                if match_pre:
                    score = float(match_pre.group(1))
                    if model == "NN":
                        MLC[dat][victim]["pre"] = score
                    elif model == "RNN":
                        MSP[dat][victim]["pre"] = score

                # Save the matched recall score
                if match_rec:
                    score = float(match_rec.group(1))
                    if model == "NN":
                        MLC[dat][victim]["rec"] = score
                    elif model == "RNN":
                        MSP[dat][victim]["rec"] = score
                
                # Save the matched f1 score
                if match_f1:
                    score = float(match_f1.group(1))
                    if model == "NN":
                        MLC[dat][victim]["f1"] = score
                    elif model == "RNN":
                        MSP[dat][victim]["f1"] = score

# Print the final dictionaries
print("MLC", MLC)
for data in MLC.keys():
    print(str(data), MLC[data])
print("MSP", MSP)
for data in MSP.keys():
    print(str(data), MSP[data])
```
